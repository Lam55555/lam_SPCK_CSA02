## Creatinine

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Nhóm dữ liệu theo nồng độ creatinine và trạng thái tử vong
creatinine_death_counts = nnbt.groupby(['serum_creatinine', 'DEATH_EVENT']).size().reset_index(name='count')

# Lọc ra số người tử vong
death_only = creatinine_death_counts[creatinine_death_counts['DEATH_EVENT'] == 1]

# Lọc ra số người không tử vong
survivor_only = creatinine_death_counts[creatinine_death_counts['DEATH_EVENT'] == 0]

# Tạo figure với 2 subplot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Vẽ biểu đồ cho người tử vong
sns.barplot(x='serum_creatinine', y='count', data=death_only, palette='viridis', ax=ax[0])
ax[0].set_title('Số người tử vong theo Nồng độ Creatinine Huyết thanh', fontsize=14)
ax[0].set_xlabel('Nồng độ Creatinine (mg/dL)', fontsize=12)
ax[0].set_ylabel('', fontsize=12)
ax[0].tick_params(axis='x', rotation=45)

# Vẽ biểu đồ cho người không tử vong
sns.barplot(x='serum_creatinine', y='count', data=survivor_only, palette='viridis', ax=ax[1])
ax[1].set_title('Số người không tử vong theo Nồng độ Creatinine Huyết thanh', fontsize=14)
ax[1].set_xlabel('Nồng độ Creatinine (mg/dL)', fontsize=12)
ax[1].set_ylabel('', fontsize=12)
ax[1].tick_params(axis='x', rotation=45)

# Tinh chỉnh layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Nhóm dữ liệu theo nồng độ creatinine và trạng thái tử vong
sodium_death_counts = nnbt.groupby(['ejection_fraction', 'DEATH_EVENT']).size().reset_index(name='count')

# Lọc ra số người tử vong
death_only = sodium_death_counts[sodium_death_counts['DEATH_EVENT'] == 1]

# Lọc ra số người không tử vong
survivor_only = sodium_death_counts[sodium_death_counts['DEATH_EVENT'] == 0]

# Tạo figure với 2 subplot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Vẽ biểu đồ cho người tử vong
sns.barplot(x='ejection_fraction', y='count', data=death_only, palette='viridis', ax=ax[0])
ax[0].set_title('Số người tử vong', fontsize=14)
ax[0].set_xlabel('Khả năng bơm máu (%)', fontsize=12)
ax[0].set_ylabel('Số ca tử vong', fontsize=12)
ax[0].tick_params(axis='x')

# Vẽ biểu đồ cho người không tử vong
sns.barplot(x='ejection_fraction', y='count', data=survivor_only, palette='viridis', ax=ax[1])
ax[1].set_title('Số người không tử vong', fontsize=14)
ax[1].set_xlabel('Khả năng bơm máu (%)', fontsize=12)
ax[1].set_ylabel('', fontsize=12)
ax[1].tick_params(axis='x')

# Thêm tiêu đề chung
fig.suptitle('Số ca tử vong và không tử vong theo khả năng bơm máu', fontsize=16)

# Tinh chỉnh layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Để tránh chồng lấn với tiêu đề chung
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Nhóm dữ liệu theo bệnh tiểu đường và trạng thái tử vong
diabetes_death_counts = nnbt.groupby(['diabetes', 'DEATH_EVENT']).size().unstack(fill_value=0)

# Tạo DataFrame chỉ với 2 cột
result_df = pd.DataFrame({
    'Số người tử vong': diabetes_death_counts[1],
    'Số người không tử vong': diabetes_death_counts[0]
})

# Vẽ biểu đồ cột với màu sắc liên quan đến bệnh tim
ax = result_df.plot(kind='bar', figsize=(10, 6), color=['#FF6F61', '#FFB3BA'])  # Màu đỏ và hồng
plt.title('Số người tử vong và không tử vong nếu có bệnh tiểu đường', fontsize=14)
plt.xlabel('Trạng thái Bệnh tiểu đường (0: Không, 1: Có)', fontsize=12)
plt.ylabel('Số người', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Trạng thái', labels=['Tử vong', 'Không tử vong'])
plt.tight_layout()
plt.show()
```

## cao huyết áp

```python
# Nhóm dữ liệu theo cao huyết áp và trạng thái tử vong
highblood_death_counts = nnbt.groupby(['high_blood_pressure', 'DEATH_EVENT']).size().unstack(fill_value=0)

# Tạo DataFrame chỉ với 2 cột
highblood_result_df = pd.DataFrame({
    'Số người tử vong': highblood_death_counts[1],
    'Số người không tử vong': highblood_death_counts[0]
})

# Vẽ biểu đồ cột với màu sắc liên quan đến bệnh tim
ax = highblood_result_df.plot(kind='bar', figsize=(10, 6), color=['#E63946', '#FF6B6B'])  # Màu đỏ đậm và hồng đậm
plt.title('Số người tử vong và không tử vong nếu bị cao huyết áp', fontsize=14)
plt.xlabel('Trạng thái Cao huyết áp (0: Không, 1: Có)', fontsize=12)
plt.ylabel('Số người', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Trạng thái', labels=['Tử vong', 'Không tử vong'])
plt.tight_layout()
plt.show()
```

```python
yes = nnbt[nnbt['DEATH_EVENT'] == 1].describe().T
no = nnbt[nnbt['DEATH_EVENT'] == 0].describe().T

colors = ['#FE251B','#FFFFFF']

fig,ax = plt.subplots(nrows = 1,ncols = 2,figsize = (8,5),)
plt.subplot(1,2,1)
sns.heatmap(yes[['mean']],annot = True,cmap = colors,linewidths = 0.4,linecolor = 'black',cbar = False,fmt = '.2f')
plt.title('Tử vong');

plt.subplot(1,2,2)
sns.heatmap(no[['mean']],annot = True,cmap = colors,linewidths = 0.4,linecolor = 'black',cbar = False,fmt = '.2f')
plt.title('Không tử vong');

fig.tight_layout(pad = -2)
plt.subplots_adjust(wspace=0.9)
plt.show()
```

## Bieu do chenh lech theo do tuoi qua cac nam

```python
# Lọc dữ liệu 'Under five' cho từng năm
under_five_2019 = df_2019[df_2019['Age'] == 'Under five']
under_five_2020 = df_2020[df_2020['Age'] == 'Under five']
under_five_2021 = df_2021[df_2021['Age'] == 'Under five']

# Gộp dữ liệu vào một DataFrame
df_under_five = pd.DataFrame({
    'Year': ['2019', '2020', '2021'],
    'Number of deaths': [under_five_2019['Number of deaths'].values[0],
                         under_five_2020['Number of deaths'].values[0],
                         under_five_2021['Number of deaths'].values[0]]
})

plt.figure(figsize=(10, 6))
plt.plot(df_under_five['Year'], df_under_five['Number of deaths'], color='gray', marker="o", linewidth=1)

# Thiết lập tiêu đề và nhãn
plt.title('Số Ca Tử Vong Nhóm dưới 5 tuổi Qua Các Năm (Biểu Đồ Đường)', fontsize=16)
plt.xlabel('Năm', fontsize=14)
plt.ylabel('Số Ca Tử Vong', fontsize=14)

# Hiển thị biểu đồ
plt.show()
```
