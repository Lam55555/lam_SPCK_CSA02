```python
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.colheader_justify', 'left')

df2 = pd.read_excel("provinces_updated.xlsx", sheet_name="Sheet1")

df2['Population'] = df2['Population'].astype(int)
df2['Area'] = df2['Area'].astype(int)

df2['Coastline'] = df2['Name'].map(df2.set_index('Name')['Coastline'])
df2['Coastline'].fillna(0, inplace=True)

df2['Coastline'] = df2['Coastline'].astype(int)

df_styled = df2.style.set_properties(**{'text-align': 'center'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'center')]}]
)

df_styled
```

```python
borders_chn = pd.read_excel('borders.xls', sheet_name='CHN')
borders_lao = pd.read_excel('borders.xls', sheet_name='LAO')
borders_khm = pd.read_excel('borders.xls', sheet_name='KHM')

borders_chn = borders_chn.rename(columns={'Tên tỉnh': 'Name'})
borders_lao = borders_lao.rename(columns={'Tên tỉnh': 'Name'})
borders_khm = borders_khm.rename(columns={'Tên tỉnh': 'Name'})

borders_chn['BorderWith'] = 'China'
borders_lao['BorderWith'] = 'Laos'
borders_khm['BorderWith'] = 'Cam'

borders_combined = pd.concat([borders_chn, borders_lao, borders_khm])

df = borders_combined.groupby('Name').agg(
    BorderWith=('BorderWith', ', '.join), 
    BorderCount=('BorderWith', 'nunique')    
).reset_index()

df_styled = df.style.set_properties(**{'text-align': 'center'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'center')]}]
)

df_styled
```
