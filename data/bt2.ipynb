## Thêm cột

```python
df['Density'] = df['Population'] / df['Area']
df
```

```python
df.drop(columns=['Density'])
```

```python
df_dense = df.groupby('Region')[['Population','Area']].sum().reset_index()
df_dense['Density'] = df_dense['Population'] / df_dense['Area']
df_dense
```

```python
df_region = df.groupby('Region').agg({
    'Area': 'sum',
    'Population': 'sum'      
}).reset_index()

df_region['Density'] = df_region['Population'] / df_region['Area']
df_region
```

## vẽ biểu đồ cột

```python
plt.figure(figsize= [13,6])
plt.bar(df_region['Region'], df_dense['Density'], color='gray', width=0.4 )
plt.title('Region', fontsize = 14)
plt.xlabel('Khu vực', fontsize = 8)
plt.ylabel('Dân số (Nghìn người)', fontsize = 8)
plt.tick_params(axis='both', labelsize=6.8)
plt.show()
```

## biểu đồ đường

```python
plt.figure(figsize= [13,6])
plt.plot(df_dense['Region'], df_dense['Density'], color='gray', marker = "o", linewidth = 1)
plt.title('Region', fontsize = 14)
plt.xlabel('Khu vực', fontsize = 8)
plt.ylabel('Dân số (Nghìn người)', fontsize = 8)
plt.tick_params(axis='both', labelsize=6.8)
plt.show()
```

## biểu đồ tròn

```python
plt.figure(figsize=[13, 6])

# Vẽ biểu đồ hình tròn
plt.pie(df_dense['Density'], labels=df_dense['Region'], autopct='%.1f%%', startangle=140)

# Thiết lập tiêu đề
plt.title('Region', fontsize=14)

# Hiển thị biểu đồ
plt.show()
```
