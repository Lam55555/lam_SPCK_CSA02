## import thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## đọc file 

```python
df = pd.read_excel(r'C:\Users\spark\CSA02\bt\tuan3\provinces_updated.xls', sheet_name="Sheet1")
df
```

## bt1

```python
df_sort_popu = df.sort_values('Population').reset_index()

df_sort_popu['Name'] = range(1, len(df_sort_popu) + 1)

# Vẽ biểu đồ cột
plt.figure(figsize=(16, 10)) 
plt.bar(df_sort_popu['Name'], df_sort_popu['Population'], color='skyblue', width=0.6) 

plt.title('Dân số các tỉnh theo thứ tự tăng dần', fontsize=16)
plt.xlabel('Tên tỉnh (Số)', fontsize=12)
plt.ylabel('Dân số', fontsize=12)
plt.xticks(df_sort_popu['Name'])

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
df_sort_popu['Index'] = range(1, len(df_sort_popu) + 1)

print('Index of provinces & cities: ', end="")
for x, y in zip(df_sort_popu['Index'], df_sort_popu['Name']):
    print(f'{x}. {y}  ', end="")
```

## region_areas.xls

```python
region_areas = pd.read_excel(r'C:\Users\spark\CSA02\bt\tuan3\region_areas.xls', sheet_name="Sheet1")
region_areas
```

## bt2

```python
plt.figure(figsize=[13, 6])
label = [ 'Đông Nam Bộ','Tây Bắc Bộ','Bắc Trung Bộ','Nam Trung Bộ','Đông Bắc Bộ','Đồng bằng sông Hồng','Đồng bằng sông Cửu Long', 'Tây Nguyên'] 
rg_ar = region_areas.set_index('Region').reindex(label).reset_index()
plt.pie(rg_ar['Area Proportion (%)'], labels=label, autopct='%.1f%%', startangle=160)

plt.title('Region', fontsize=14)

plt.show()
```

## bt3

```python
population_density = df['PopulationDensity']

plt.figure(figsize=[10,10]) 
plt.xticks(range(50, int(population_density.max()) + 1000, 230))
plt.hist(population_density, bins=20, color='skyblue', edgecolor='black') 

plt.title('Frequency of Population Density in VietNam Provinces & Cities', fontsize=14)
plt.xlabel('Population Density (People / km2)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
```
